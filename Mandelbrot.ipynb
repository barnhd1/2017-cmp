```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def mandelbrot_maker(top_num, threshold, y_axis, x_axis):
    x = np.linspace(-2, 1, x_axis)
    y = np.linspace(-1.5, 1.5, y_axis)
    
    c = x[:,np.newaxis] + 1j*y[np.newaxis,:]
    
    z = c

    for i in range(top_num):

        z = z * z + c
        
    mandlebrot_set = (abs(z) < threshold)
    
    return mandlebrot_set

mandelbrot_set = mandelbrot_maker(1000, 1000, 601, 601)
plt.figure(figsize=(9,9), dpi=200)

plt.imshow(mandelbrot_set.T,cmap = "Greys" )
```
